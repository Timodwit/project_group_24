```python
# Importeren van dingen
import numpy as np
import matplotlib.pyplot as plt
import math

# L_ext functie van F
k_boltz = 1.381 * (10 ** (-23))

def Func_L(L_ext, T, L_c, L_p):

    return (k_boltz * T / L_p) * (1 / (4*((1 - L_ext / L_c) ** 2)) - 1 / 4 + L_ext / L_c)

x_L = np.arange(0, 10000 * 0.50 * (10 ** (-9)), 10000 * 0.01 * (10 ** (-9)))

#x_L = np.arange(0, 50 , 0.1 )

y_L = []
bijhouden_n = 0
L_ext2 = []
F_variatie = [0.1 * (10 ** (-12)), 0.2 * (10 ** (-12)), 0.3 * (10 ** (-12)), 0.4 * (10 ** (-12)), 0.8 * (10 ** (-12)), 1 * (10 ** (-12)), 3 * (10 ** (-12)), 5 * (10 ** (-12)), 6 * (10 ** (-12))]
for i in range(len(x_L)):
    y_L.append(Func_L(x_L[i], 293, 10000 * 0.34 * (10 ** (- 9)), 45 * (10 ** (-9))))

    if bijhouden_n == len(F_variatie):
        bijhouden_n == bijhouden_n

    else:
        if y_L[i] >= F_variatie[bijhouden_n]:
            bijhouden_n = bijhouden_n + 1
            L_ext2.append(x_L[i])


#Func_L(x_L, 293, 10000 * 0.34 * (10 ** (- 9)), 45 * (10 ** (-9)))
plt.scatter(x_L, y_L)
plt.show()
```
